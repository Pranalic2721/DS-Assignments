#2. What is the difference between the Label encoding and One-hot encoding?
Ans:-

-> Label encoding assigns integers to categories, but can imply false ordinal relationships.

-> One-hot encoding creates binary columns for each category, avoiding implied ordering.

->Label encoding is simple to do and keeps the order of the categories, which may be advantageous in some cases.

->One−shot coding is a common approach for transforming categorical variables into numeric values.
