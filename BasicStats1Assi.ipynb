Data Visualization


```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
data=pd.read_csv("/content/sales_data_with_discounts.csv")
data
```

```python
data.shape
```

```python
data.Volume.value_counts().plot(kind="pie")
```

```python
data.Volume.value_counts().plot(kind="bar")
```

```python
data.Volume.value_counts().plot(kind="hist")
```

```python
plt.scatter(data.SKU,data.Volume,color="red")
```

```python
plt.hist(data.Volume,edgecolor="red")
```

```python
plt.boxplot(data.Volume)
plt.show()
```

```python
plt.boxplot(data.Volume,vert=False,notch=True)
plt.show()
```

```python
plt.figure(facecolor='yellow')
plt.scatter(data.SKU,data.Volume,marker='*')
plt.show()
```

```python
data.hist(figsize=(15,5))
plt.tight_layout()
plt.show()
```

```python
plt.violinplot(data["Volume"])
plt.show()
```
